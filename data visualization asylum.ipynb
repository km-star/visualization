# For Pie-chart Visualization:


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
data=pd.read_csv("C:\\Users\\Dell\\asylum_seekers_monthly.xls")
y=data.Origin.value_counts()
val=[]
sta=[]
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
for x in range(0,len(y),1):
    if(y[x]>=30):
         val.append(y[x])
key=y.keys()
for i in range(0,len(val),1):
 sta.append(key[i])
ax.pie(val,labels=sta,autopct='%1.2f%%')
fig.set_size_inches(18.5, 10.5)
```

# For Scatter-Plot Visualization:


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
ASM=pd.read_csv("C:\\Users\\Dell\\asylum_seekers_monthly.xls")
df = pd.DataFrame(ASM)
df['Value'] = pd.to_numeric(df['Value'],errors='coerce')
df = df.replace(np.nan, 0, regex=True)
df['Value'] = df['Value'].astype(int)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
x=df["Year"]
y=df["Value"]
ax.set_xlabel("Year",fontsize="20")
ax.set_ylabel("Value",fontsize="20")
ax.set_title("Immigrants accordind to year between 1999 to 2016",fontsize="20"
)
ax.scatter(x,y)
fig.set_size_inches(15,10)
```

# For Heatmap Visualization:

```python
#data preprocessing

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

ASM= pd.read_csv("C:\\Users\\Dell\\asylum_seekers_monthly.xls")
df = pd.DataFrame(ASM)
df.isna().sum()
df[df.Value=='*']
df=df[df.Value!='*']
```

```python
df['Value']=pd.to_numeric(df['Value'])
```

```python
df.info()
```

```python
#data visualization

grouped_df =df.groupby(['territory_of_asylum','Origin'])['Year'].count().reset_index()
grouped_df = grouped_df.pivot('territory_of_asylum', 'Origin', 'Year')
grouped_df.head()
plt.figure(figsize=(20,20))
sns.heatmap(grouped_df,cmap="YlGnBu")
plt.title("Heatmap showing territory_of_asylum vs Origin in yearwise")
plt.show()
```
